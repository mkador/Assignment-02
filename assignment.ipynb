### 1. Email Validator (String Filtering + Loops)
- Real-world use: Used in signup forms or data cleaning tasks.
Task:
Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.

- Example:
Input: hello@inceptionbd.com
Output: Valid Email

```python
email = input("Enter an email: ")

flag = False

for i in email:
    if i == "@":
        flag = True

if flag and (email[-4:] == ".com" or email[-4:] == ".org"):
    print('Valid Email')
else:
    print("Invalid Email")
```

### 2. Count Word Frequency in Customer Reviews
- Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
- Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
text = input("Enter the text: ").lower()

words = text.split()   
count_word = {}          

for w in words:
    if w in count_word:
        count_word[w] += 1
    else:
        count_word[w] = 1

print(count_word)
```

### 3. Password Strength Checker
üíº Real-world use: Used in user authentication systems.
Task:
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.
Example:
Input: "Hello@123"
Output: Strong Password

```python
password = input("Enter a password: ")

has_upper = False
has_lower = False
has_digit = False
has_special = False

for pw in password:
    if pw.isupper():
        has_upper = True
    elif pw.islower():
        has_lower = True
    elif pw.isdigit():
        has_digit = True
    else:
        has_special = True

if has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

### 4. Username Generator
üíº Real-world use: Automatically create usernames for users.
Task:
Take user‚Äôs full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).
Example:
Input: "Mamun Malitha"
Output: mamitha57

```python
name = input("Enter full name: ").lower()
parts = name.split()

first_name = parts[0]         
last_name = parts[-1]          


first_3letter = ""
for i in range(3):
    if i < len(first_name):
        first_3letter += first_name[i]


last_3letter = ""
start = len(last_name) - 3
if start < 0:
    start = 0

for i in range(start, len(last_name)):
    last_3letter += last_name[i]


total = 0
for ch in name:
    total += ord(ch)

number = total % 100

print(first_3letter + last_3letter + str(number))
```

### 5. SMS Word Counter
üíº Real-world use: Used in messaging apps or social media post limit check.
Task:
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).
Example:
Input: "This is a demo message."
Output: Words = 5, Characters = 23, Status = Within Limit

```python

message = input("Enter your message: ")

words = message.split()
word_count = len(words)

char_count = len(message)

if char_count > 160:
    status = "Exceeded Limit"
else:
    status = "Within Limit"

print(f"Words = {word_count}, Characters = {char_count}, Status = {status}")
```

### 6. Remove Duplicates from Contact List
üíº Real-world use: Data cleaning in CRM or user databases.
Task:
Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

```python
n_input = input("Enter names with comma: ")

names_list = n_input.split(',')
unique_names = []
for name in names_list:
    name = name.strip()
    if name not in unique_names:
        unique_names.append(name)

print(", ".join(unique_names))
```

### 7. Detect Suspicious Words in Text (Moderation)
üíº Real-world use: Used in chat moderation or email filtering.
- Task:
If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.
- Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected

```python
text = input("Enter text: ").lower()

banned_word = ['spam', 'scam', 'fake']

warning = False

for word in banned_word:
    if word in text:
        warning = True
        break

if warning:
    print("Warning: Suspicious content detected")
else:
    print("Safe: No suspicious words found")
```

### 8. Invoice Code Formatter
üíº Real-world use: Formatting IDs or codes in finance or ERP systems.
- Task:
Given an invoice number like 23, generate code in the format INV00023.
- Example:
Input: 23
Output: INV00023

```python
num = int(input("Enter invoice number: "))
invoice_formatted = str(num)
while len(invoice_formatted) < 5:
    invoice_formatted = "0" + invoice_formatted
invoice_code = "INV" + invoice_formatted

print(invoice_code)
```

### 9. Count Hashtags and Mentions (Social Media Analytics)
üíº Real-world use: Used in analyzing social media posts.
- Task:
Count how many hashtags (#) and mentions (@) appear in a post.
- Example:
Input: "Check out #AIWorkshop and follow @inceptionbd"
Output: Hashtags = 1, Mentions = 1

```python
post = input("Enter your post: ")

hashtag_count = 0
mention_count = 0

for ch in post:
    if ch == '#':
        hashtag_count += 1
    elif ch == '@':
        mention_count += 1

print("Hashtags =", hashtag_count, ", Mentions =", mention_count)
```

### 10. Word Reversal for Data Encryption
üíº Real-world use: Simple text obfuscation or playful encoding.
- Task:
Reverse each word in a sentence while keeping word order the same.
- Example:
Input: "AI is fun"
Output: "IA si nuf"

```python
text = input("Enter a text: ")

words = text.split()    
reversed_words = []         

for w in words:
    reversed_words.append(w[::-1])  

output = " ".join(reversed_words)    

print(output)
```
